```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\Assignment\Clustering\crime_data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
trans= MinMaxScaler()
df_norm= pd.DataFrame(trans.fit_transform(data.iloc[:,1:]))
df_norm
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.45, min_samples=5)
dbscan.fit(df_norm)         
```

```python
#Noisy samples are given the label -1.
dbscan.labels_ 
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
data_1= pd.concat([data,cl],axis=1)
data_1
```

```python
data_1.plot(x="Murder",y ="Assault",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')   
```

```python
dl = dbscan.labels_ 
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df_norm, dl)
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(df_norm) 
y_kmeans 
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['Kcluster']) 
cl1
```

```python
data_2= pd.concat([data,cl1],axis=1) 
data_2
```

```python
data_2.plot(x="Murder",y ="Assault",c=y_kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans')   
```

```python
silhouette_score(df_norm,y_kmeans)# KMeans cluster slighty improves the accuracy
```
